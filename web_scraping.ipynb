```python
for tag in soup.find_all('a'):
    print(tag.get('href'))
```

```python
links=[]
for a in soup.findAll('a',href=True):
    if re.findall('npc', a['href']):
        links.append(a['href'])  
```

```python
range(len(links))
```

```python
names=[]
for i in range(len(links)):
    names.append(links[i][5:-5])
```

```python
count = 0
names_dic = {} #Empty dictionary to add values into
for i in names:
    names_dic[count] = i
    count+=1
len(names_dic )   
names_dic
```

```python
What_is_this_product_used_for=[]
Last_Reviewed_Date=[]


for i in range(len(names_dic)):
    page="https://www.drugs.com/npc/"+names_dic[i]+".html"
    pages = requests.get(page)
    Products = BeautifulSoup(pages.content, 'html.parser')
    heading= Products.select('h2')
    headings=[]
    for i in heading:
        headings.append(i.text)
    headings
    if ('What is this product used for?' in headings):
        ourList = Products.find(attrs={"class":"contentBox"})       
        product_used=ourList.find_all('div')[headings.index('What is this product used for?')+1].text
        product_used = product_used.replace('\n', ' ')
    else: 
        product_used="none"
    if ('Last Reviewed Date' in headings):
        Reviewed_Date=ourList.find_all('div')[headings.index("Last Reviewed Date")+1].text
        Reviewed_Date = Reviewed_Date.replace('\n', ' ')
    else: 
        Reviewed_Date="none"
    What_is_this_product_used_for.append(product_used)
    Last_Reviewed_Date.append(Reviewed_Date)
        
        
        

Last_Reviewed_Date

```

```python
output=pd.DataFrame({"names":names_dic.values() ,
                     "product_used" :What_is_this_product_used_for, 
                     'Last_Reviewed_Date':Last_Reviewed_Date})
output
```

```python
output.to_csv('output.csv')
```
